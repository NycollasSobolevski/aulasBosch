1) Represente graficamente as funções $\sinh (x)$, $\cosh(x)$ e $\tanh(x)$ no intervalo $-3 \leq x \leq 3$. Utilize cores diferentes para cada curva e inclua uma legenda, identificando cada função. O vetor $x$ deve ser um ```NumPy``` array com, digamos, 100 valores linearmente espaçados no intervalo.

````python
import numpy as np
import matplotlib.pyplot as plt
import math

a = np.linspace(-3, 3, 100)
aux = list(a)
s = []
c = []
t = []
for i in range(len(aux)):
    s.append(math.sin(aux[i]))
    c.append(math.cos(aux[i]))
    t.append(math.tan(aux[i]))

fig, eixo = plt.subplots(nrows=1, ncols=2, figsize = (20,5))    
eixo[0].plot(aux, s, label = "Seno", color = "r")
eixo[0].legend()
eixo[1].plot(aux, c, label = "Cosseno")
eixo[1].legend()
plt.show()

plt.figure(figsize=(20,5))
plt.plot(aux, t, label = "Tangente")
plt.legend()
plt.show()
````

2) Elabore um gráfico do tipo pizza contendo o número de habitantes de 6 cidades da Região Metropolitana de Curitiba de sua escolha (consultar os dados em https://pt.wikipedia.org/wiki/Regi%C3%A3o_Metropolitana_de_Curitiba).

````python
import numpy as np
import matplotlib.pyplot as plt
import math

labels = ["Araucária", "Campo Largo", "Colombo", "Fazenda Rio Grande", "Pinhais", "São José dos Pinhais"]
val = [141410, 130091, 240840, 98368, 130789, 317476]

fig1,ax1 = plt.subplots(figsize=(8,8))

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d} hab.\n {p:.2f}%'.format(v = val,p = pct)
    return my_autopct

print(make_autopct(val))

ax1.pie(val, labels=labels, autopct = make_autopct(val))
plt.show()
````

3) O código abaixo produz uma série de pontos distribuídos de forma randômica no intervalo $0 \leq x \leq 20$, e também um gráfico da distribuição. Seu trabalho é acrescentar a esse mesmo gráfico três linhas horizontais:



1.   Uma linha representando a média da distribuição de pontos (em azul, linha preenchida);
2.   Uma linha com o valor (media+desvio_padrao), em azul, traço-traço.
3.   Uma linha com o valor (media-desvio_padrao), em azul, traço-traço.



````python
import matplotlib.pyplot as plt
import numpy as np
# criação da distribuição de pontos
N = 100
x = np.linspace(0,20,N)
y = np.sin(x)*np.cos(x)*np.random.rand(N) - 0.5*np.random.rand(N)

# gráfico de y em função de x
plt.plot(x,y,'k-')
plt.xlabel('x'); plt.xlim([0,20])
plt.ylabel('y'); plt.ylim([-1,0.5])

# cálculo da média e do desvio padrão com funções prontas do NumPy
media = np.mean(y)

m = [media]*100
plt.plot (m, linestyle="-",
          color= "b", linewidth=2.0)

desvio_padrao = np.std(y)
l1 = media+desvio_padrao
l2 = media-desvio_padrao

ls = [l1]*100
li = [l2]*100

plt.plot (ls, linestyle="--",
          color= "b", linewidth=1.5)
plt.plot (li, linestyle="--",
          color= "b", linewidth=1.5)
````

4) O programa abaixo contém listas de temperatura máxima média e mínima média para a cidade de Curitiba ao longo do ano. Elabore um gráfico preenchido (ver ```plt.fill_between```) com essas temperaturas em função dos meses do ano. Para referência, seu gráfico deve ficar parecido com aquele mostrado neste link: https://weather-and-climate.com/uploads/average-temperature-brazil-curitiba.png

````python
import matplotlib.pyplot as plt
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
# temperaturas máxima média e mínima média (em graus Celsius)
temp_max = [26.8,	26.8,	26,	24,	20.8,	20.1,	19.7, 21.5,	21.4,	23.1,	25,	26.2]
temp_min = [17.2,	17.4,	16.5,	14.6,	11.2,	9.7, 9, 9.6, 11.1,	13.2,	14.9,	16.2]

fig1,ax1 = plt.subplots(figsize=(10,4))

# continue a solução 
ax1.plot (meses, temp_max, linestyle="-",
          color= "r", marker ="o", linewidth=1.5, label = "Max temp")
ax1.plot (meses, temp_min, linestyle="-", label = "Min temp",
          color= "b", marker ="o", linewidth=1.5)

ax1.fill_between(meses, temp_max, temp_min, color ="indianred")
ax1.fill_between(meses, temp_min, 0, color ="royalblue")
plt.yticks([0, 15, 30])
plt.axis([0,11,0,30])

plt.xlabel("Meses")
plt.ylabel("Temperatura")
plt.grid(True, linewidth=1, linestyle="-")
````
